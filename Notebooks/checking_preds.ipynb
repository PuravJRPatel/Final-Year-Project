```python
import plotly.express as px
import pandas as pd
import torch

# Load predicted embeddings
predicted_2024 = torch.load("graph_embeddings/embeddings_2024.pt").detach().numpy()

# Load the original dataset for geographical information
data_2023 = pd.read_csv(r"C:\Users\purav\OneDrive\Desktop\Fi Year Project\Final-Year-Project\Data\Processed\climate_data.csv")
data_2023 = data_2023[data_2023['year'] == 2023].reset_index(drop=True)

# Add predictions to the DataFrame
data_2023['predicted_rainfall'] = predicted_2024[:, 0]
data_2023['predicted_temperature'] = predicted_2024[:, 1]

# Visualize Rainfall Predictions
fig_rainfall = px.scatter_mapbox(
    data_2023,
    lat='latitude',
    lon='longitude',
    color='predicted_rainfall',
    size='predicted_rainfall',
    color_continuous_scale='viridis',
    title='Predicted Rainfall for 2024',
    mapbox_style='carto-positron',
    zoom=4
)
fig_rainfall.show()

# Visualize Temperature Predictions
fig_temperature = px.scatter_mapbox(
    data_2023,
    lat='latitude',
    lon='longitude',
    color='predicted_temperature',
    size='predicted_temperature',
    color_continuous_scale='plasma',
    title='Predicted Temperature for 2024',
    mapbox_style='carto-positron',
    zoom=4
)
fig_temperature.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the merged dataset
merged_file = r"C:\Users\purav\OneDrive\Desktop\Fi Year Project\Final-Year-Project\Data\Processed\grid_with_data.geojson"
merged_gdf = gpd.read_file(merged_file)

# Choose a year to visualize
year = 2020  # Change this as needed

# Plot rainfall data
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

merged_gdf.plot(column=f"rainfall_{year}", cmap="Blues", legend=True, ax=axes[0])
axes[0].set_title(f"Rainfall - {year}")

# Plot temperature data
merged_gdf.plot(column=f"temperature_{year}", cmap="Reds", legend=True, ax=axes[1])
axes[1].set_title(f"Temperature - {year}")

plt.tight_layout()
plt.show()
```
